```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt

# Path to YOLO-predicted images
image_dir = 'runs/detect/wafer_model12_test_eval3/'
image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))]

# Randomly sample 100 images
sampled_files = random.sample(image_files, 100)

# Load and resize images for consistent layout
images = [Image.open(os.path.join(image_dir, f)).resize((128, 128)) for f in sampled_files]

# Create a grid (10x10)
rows, cols = 10, 10
fig, axes = plt.subplots(rows, cols, figsize=(20, 20))

for ax, img in zip(axes.flatten(), images):
    ax.imshow(img)
    ax.axis('off')

# Hide any unused axes
for ax in axes.flatten()[len(images):]:
    ax.axis('off')

plt.tight_layout()
plt.savefig('compound_grid.png')  # Save the result
plt.show()
```
